```python
def type_M(value):
    """функция принимает 1 аргумент value - минерализация воды (г/л) и возвращает название группы, к которой вода относится по минерализации
    (Петров Н.С., Потапов А.А. Общая гидрогеология. Практикум. - СПб., 1993 с.45)"""
    water, type = '', '' 
    if value <= 35:
        water = 'воды'
        if value <= 0.2:
            type = 'ультрапресные'
        elif 0.2 < value <= 1:
            type = 'пресные'
        elif 1 < value <= 3:
            type = 'слабосолоноватые'
        elif 3 < value <= 10:
            type = 'солоноватые'
        elif 10 < value <= 35:
            type = 'соленые'
    else:
        water = 'рассолы'
        if 35 < value <= 100:
            type = 'слабые'
        elif value > 100:
            type = 'крепкие'
    return type + ' ' + water
```

```python
def type_pH(value):
    """функция возвращает название группы подземных минеральных вод по величине pH (А.Н.Павлов, В.Н. Шемякин, 1967 г.)"""
    water, type = 'воды', ''
    if value <= 0 or value > 14:
        return 'некорректное значение величины pH' 
    if value < 1.9:
        type = 'сильнокислые'
    elif 1.9 <= value < 4.1:
        type = 'кислые'
    elif 4.1 <= value < 7:
        type = 'слабокислые'
    elif value == 7:
        type = 'нейтральные'
    elif 7 < value < 8.3:
        type = 'слабощелочные'
    elif 8.3 <= value < 10.3:
        type = 'щелочные'
    elif value > 10.3:
        type = 'сильнощелочные'
    if type:
        return water + ' ' + type
```

```python
type_M(0.8)
```

```python
type_pH(6.9)
```
